```python
#prstSi
sim_no = 5 # number of simulations/seeds

ps2s = 1e-12 

VACFs_prstSi = []
for i in range(1,sim_no+1):

    filename_prstSi = 'vacf_all_Pristine_Silicon_200.d'
    VACFs_prstSi.append(np.loadtxt(filename_prstSi))
VACFs_prstSi = np.array(VACFs_prstSi)
print(np.shape(VACFs_prstSi))

mean_vacf_all_prstSi = np.mean(VACFs_prstSi, axis = 0)
print(np.shape(mean_vacf_all_prstSi))

dt_sys_prstSi = 0.0005
frequency_prstSi = 20
number_of_points_prstSi = 1000

VDOS_prstSi = []
for i in range(0,sim_no):#

    t_all_prstSi = VACFs_prstSi[i,:,0]*dt_sys_prstSi
    c_all_prstSi = VACFs_prstSi[i,:,-1]
    
    freq_prstSi = np.linspace(0,frequency_prstSi,number_of_points_prstSi)

    VDOS_sys_prstSi = []   
    for f in freq_prstSi:
        transform_prstSi = np.dot(c_all_prstSi,np.cos(2*np.pi*t_all_prstSi*f))*(dt_sys_prstSi)
        VDOS_sys_prstSi.append(np.sqrt(transform_prstSi**2))
    t_all_prstSi = []; c_all_prstSi = [] 

#    #VDOS_sys = np.array(VDOS_sys)
    VDOS_prstSi.append(VDOS_sys_prstSi)
VDOS_prstSi = np.array(VDOS_prstSi)    

#print(np.shape(VDOS_prstSi))


mean_vdos_prstSi = np.mean(VDOS_prstSi, axis = 0)
print(np.shape(mean_vdos_prstSi))
mean_vdos_prstSi = np.array(mean_vdos_prstSi)

from scipy.ndimage import gaussian_filter
    
smooth_VDOS_prstSi = gaussian_filter(VDOS_prstSi, sigma=3)

smooth_VDOS_all_prstSi = gaussian_filter(mean_vdos_prstSi, sigma=3)
```
